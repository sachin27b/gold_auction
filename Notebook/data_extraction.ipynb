```python
import os
import pandas as pd
from tqdm import tqdm
import camelot

def read_pdf(directory: str):
    # Iterate over all files in the given directory
    for filename in os.listdir(directory):
        if filename.endswith(".pdf"):
            pdf_path = os.path.join(directory, filename)
            print(f"Processing {filename}...")
            all_tables = pd.DataFrame()

            # Get the number of pages in the PDF using camelot to extract from all pages
            all_page_tables_list = camelot.read_pdf(pdf_path, pages='all', strip_text='\n')
            for table in tqdm(all_page_tables_list):
                tab = table.df
                tab.columns = [None] * len(tab.columns)
                tab = tab.iloc[1:-1]  # Skip header and footer rows
                all_tables = pd.concat([all_tables, tab], axis=0, ignore_index=True)

            # Drop the 9th column if it exists
            if all_tables.shape[1] > 8:
                all_tables = pd.concat([all_tables.iloc[:, :8], all_tables.iloc[:, 9:]], axis=1)

            # Rename columns
            all_tables.columns = ['Agreement Number', 'Disbursement Date', 'Loan Amount', 'Number of Ornaments', 
                                  'Gross Auction Weight of Ornaments(gm)', 'Auction Date', 'Auction Place', 
                                  'Branch Address']

            # Save the DataFrame as CSV
            csv_filename = os.path.splitext(filename)[0] + '.csv'
            all_tables.to_csv(csv_filename, index=False)
            print(f"Saved {csv_filename}")
```

```python
read_pdf('../Data/')
```

```python


# Specify the directory containing the CSV files
directory = '../Data/'

# Initialize an empty list to store DataFrames
dataframes = []

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        # Read the CSV file
        df = pd.read_csv(os.path.join(directory, filename))
        # Append the DataFrame to the list
        dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
df_data = pd.concat(dataframes, ignore_index=True)


```
